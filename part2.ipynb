__4. Set up initial model__

```python
# Shift variables that cannot be known in advance forward in time by 2 days
time_sensitive_features = ['TABLET_REG','DESKTOP_REG','MOBILE_REG','TV_REG','UPLOADERS','WEB_FREE_TRIALS','ANDROID_FREE_TRIALS','IOS_FREE_TRIALS']
target_col = 'FTS'

for feature in time_sensitive_features:
    new_feature = feature +'_1days_ago'
    #print(new_feature)
    data_input.loc[:,new_feature] = data_input.loc[:,feature].shift(1) #shift here is the number of rows you want to shift over
```

```python
data_input.head()
```

```python
data_input = data_input.iloc[1:]
```

```python
data_input.shape
```

```python
X = data_input.drop(target_col, axis = 'columns')
y = data_input.loc[:,target_col]
```

```python
#refine the feature columns
feature_cols = []
for column in X:
    if column.endswith('_1days_ago'):
        feature_cols.append(column)
```

```python
feature_cols.append('SEASON_NUM')
feature_cols.append('IS_HOLIDAY')
feature_cols.append('IS_FRI')
feature_cols.append('IS_SAT')
feature_cols.append('IS_SUN')
feature_cols.append('IS_MON')
feature_cols.append('IS_TUE')
feature_cols.append('IS_WED')
feature_cols.append('IS_THU')
```

```python
X = X.loc[:,feature_cols]
X.head()
```

```python
#Split your rows into a training set and a test set.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,shuffle=False, random_state = 16)
```

```python
# Confirm that the split worked as expected
print(X_train.index)
print(X_test.index) 
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

predictions = lr.predict(X_test)
mse = mean_squared_error(y_test,predictions)
rmse = np.sqrt(mse)
rmse
```

```python
lr.score(X_test, y_test)
```

```python
list(zip(lr.predict(X_test), y_test))
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
print(lr.intercept_)
print(list(zip(X.columns, lr.coef_)))
```
